# Analísis descriptivo: ***Relación entre Educación , Género , Tecnología y Características del Hogar***


---


1.   **Educación, Género y Tecnología**

      El análisis muestra cómo la equidad de género, el acceso a internet y la disponibilidad tecnológica interactúan en los hogares:

      Equidad de Género: La distribución de género en los hogares refleja que una proporción significativa presenta equilibrio entre hombres y mujeres, mientras que los hogares con predominio de hombres o mujeres son menos frecuentes.

      Impacto de la Educación: Variables como escolaridad_mejora, educacion_empleo y educacion_decisiones indican que la tecnología tiene un impacto positivo en la educación y las decisiones tomadas en los hogares, con promedios que sugieren una percepción de mejora en estas áreas.

      Disponibilidad Tecnológica: Los dispositivos como computadoras, televisores digitales y smartphones están presentes en la mayoría de los hogares, pero el acceso a tecnologías más avanzadas, como tablets, es menos frecuente.

2.   **Disponibilidad de Tecnología y Características del Hogar**

    El análisis de las tecnologías disponibles en función del tamaño del hogar y sus características proporciona información valiosa:

    Disponibilidad de Tecnologías por Tamaño del Hogar: En general, los hogares con más integrantes tienden a tener una mayor disponibilidad promedio de dispositivos tecnológicos. Esto podría deberse a la necesidad compartida de acceso o la capacidad económica de hogares más grandes.

    Distribución de Género por Tamaño del Hogar: A medida que aumenta el tamaño del hogar, también crece el promedio de hombres y mujeres. Esto refleja que los hogares grandes suelen estar compuestos por familias completas con mayor paridad de género.

    Proporción de Personas Menores de 30 Años: Los hogares más grandes tienen una mayor proporción de personas menores de 30 años. Esto sugiere que los hogares con más integrantes tienden a ser más jóvenes, posiblemente debido a la presencia de niños o adultos jóvenes.

```python
import matplotlib.pyplot as plt

# Resumen estadístico para variables relacionadas con educación y tecnología
education_tech_columns = [
    'escolaridad_mejora', 'educacion_empleo', 'educacion_decisiones',
    'compu_escritorio', 'compu_portatil', 'tv_digital',
    'tablet_disponible', 'celular_smartphone'
]

education_tech_stats = dataset[education_tech_columns].describe()

# Distribución de equidad de género
gender_distribution = dataset['equidad_genero'].value_counts(normalize=True) * 100

# Internet disponibilidad por equidad de género
internet_by_gender = dataset.groupby('equidad_genero')['internet_disponibilidad'].value_counts(normalize=True).unstack()

# Crear gráficos para el análisis descriptivo
# 1. Distribución de Equidad de Género
plt.figure(figsize=(8, 5))
gender_distribution.plot(kind='bar', title='Distribución de Equidad de Género (%)')
plt.xlabel('Equidad de Género')
plt.ylabel('Porcentaje')
plt.show()

# 2. Disponibilidad de Internet por Género
internet_by_gender.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Disponibilidad de Internet según Equidad de Género')
plt.xlabel('Equidad de Género')
plt.ylabel('Proporción')
plt.legend(title='Disponibilidad de Internet')
plt.show()

education_tech_stats
```

**Impacto** de la Tecnología en la Educación:


---


Las estadísticas indican que, en promedio, los hogares reportan una percepción de mejora en la escolaridad (escolaridad_mejora con una media de 1.36) debido al acceso a la tecnología. Esto sugiere que los dispositivos tecnológicos pueden desempeñar un papel clave en el aprendizaje y la adquisición de habilidades.


**Relación entre Educación y Empleo:**

    La variable educacion_empleo muestra un promedio de 1.41, lo que implica que los hogares reconocen una conexión entre la educación y el acceso a mejores oportunidades laborales. Esto destaca la importancia de la tecnología como herramienta de inclusión en el ámbito profesional.

**Educación y Toma de Decisiones:**

    Con una media de 1.66 en la variable educacion_decisiones, se observa que la tecnología influye positivamente en la capacidad de los individuos para tomar decisiones informadas en los hogares.

**Disponibilidad de Dispositivos Tecnológicos:**

    Los datos reflejan una alta presencia de dispositivos como computadoras de escritorio (media de 1.77), portátiles (1.53) y smartphones (1.04). Sin embargo, tecnologías como tablets tienen una menor disponibilidad (1.77), lo que podría representar una brecha en el acceso equitativo a herramientas tecnológicas más modernas.

**Limitaciones en el Acceso Avanzado:**texto en negrita**

    A pesar de la presencia de tecnologías básicas en la mayoría de los hogares, la variabilidad en el acceso a dispositivos más avanzados resalta desigualdades que podrían estar relacionadas con factores económicos o demográficos.

**Recomendación General:**

    Se sugiere fomentar políticas públicas y programas que reduzcan las barreras para acceder a tecnologías avanzadas en los hogares, dado su impacto positivo en la educación y la inclusión laboral.

```python
import matplotlib.pyplot as plt

# Resumen estadístico para variables relacionadas con educación y tecnología
education_tech_columns = [
    'escolaridad_mejora', 'educacion_empleo', 'educacion_decisiones',
    'compu_escritorio', 'compu_portatil', 'tv_digital',
    'tablet_disponible', 'celular_smartphone'
]

education_tech_stats = dataset[education_tech_columns].describe()

# Distribución de equidad de género
gender_distribution = dataset['equidad_genero'].value_counts(normalize=True) * 100

# Internet disponibilidad por equidad de género
internet_by_gender = dataset.groupby('equidad_genero')['internet_disponibilidad'].value_counts(normalize=True).unstack()

# Crear gráficos para el análisis descriptivo
# 1. Distribución de Equidad de Género
plt.figure(figsize=(8, 5))
gender_distribution.plot(kind='bar', title='Distribución de Equidad de Género (%)')
plt.xlabel('Equidad de Género')
plt.ylabel('Porcentaje')
plt.show()

# 2. Disponibilidad de Internet por Género
internet_by_gender.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Disponibilidad de Internet según Equidad de Género')
plt.xlabel('Equidad de Género')
plt.ylabel('Proporción')
plt.legend(title='Disponibilidad de Internet')
plt.show()


education_tech_stats
```

```python
# Relación entre disponibilidad de tecnología y características del hogar
tech_columns = [
    'compu_escritorio', 'compu_portatil', 'tv_digital', 'tablet_disponible', 'celular_smartphone'
]
home_columns = ['personas_vivienda', 'hombres_vivienda', 'mujeres_vivienda', 'personas_0_29']

# Calcular promedios de tecnologías según el tamaño del hogar y composición
tech_availability_by_home_size = dataset.groupby('personas_vivienda')[tech_columns].mean()
gender_distribution_by_tech = dataset.groupby('personas_vivienda')[['hombres_vivienda', 'mujeres_vivienda']].mean()
age_distribution_by_tech = dataset.groupby('personas_vivienda')['personas_0_29'].mean()

# Graficar los resultados
# 1. Disponibilidad de tecnologías por tamaño del hogar
plt.figure(figsize=(12, 6))
tech_availability_by_home_size.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Disponibilidad de Tecnologías según el Tamaño del Hogar')
plt.xlabel('Número de Personas en el Hogar')
plt.ylabel('Promedio de Disponibilidad')
plt.legend(title='Tecnologías', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()

# 2. Distribución de género por tamaño del hogar
plt.figure(figsize=(10, 5))
gender_distribution_by_tech.plot(kind='bar', figsize=(10, 6))
plt.title('Distribución de Género según el Tamaño del Hogar')
plt.xlabel('Número de Personas en el Hogar')
plt.ylabel('Promedio de Hombres y Mujeres')
plt.legend(title='Género', loc='upper left')
plt.grid(axis='y')
plt.show()

# 3. Edad promedio (menores de 30 años) por tamaño del hogar
plt.figure(figsize=(10, 5))
age_distribution_by_tech.plot(kind='line', marker='o', color='green', figsize=(10, 6))
plt.title('Proporción de Personas Menores de 30 años según el Tamaño del Hogar')
plt.xlabel('Número de Personas en el Hogar')
plt.ylabel('Promedio de Personas 0-29 Años')
plt.grid()
plt.show()
```
