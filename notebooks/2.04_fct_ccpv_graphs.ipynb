# Graph Beta

```python
# First we filter to register totals for each Federal entity:
df_2000=df_2000[( ( df_2000['Nombre del municipio o delegación'].str.contains("Total de la entidad", na=False) )|( df_2000['Nombre del municipio o delegación'].str.contains("Total del Distrito", na=False) ))
                 &(df_2000['Nombre de la localidad']=='TOTAL DE LA ENTIDAD')]

df_2005=df_2005[( (df_2005['Municipio'].str.contains("Total de la entidad", na=False) ) |( df_2005['Municipio'].str.contains("Total del Distrito", na=False) ))
                &(df_2005['Localidad']=='Total de la Entidad')]

df_2010=df_2010[( (df_2010['Municipio ó delegación'].str.contains("Total de la entidad", na=False))|( df_2010['Municipio ó delegación'].str.contains("Total del Distrito", na=False) ))
                &(df_2010['Localidad']=='Total de la Entidad')]

df_2020=df_2020[(df_2020['Municipio o demarcación territorial'].str.contains("Total de la entidad", na=False))&(df_2020['Localidad']=='Total de la Entidad')]
```

```python

# Federal entity
df_2000['Entity']=df_2000['Nombre de la entidad']
df_2005['Entity']=df_2005['Entidad Federativa']
df_2010['Entity']=df_2010['Entidad federativa']
df_2020['Entity']=df_2020['Entidad federativa']

# Average grade
df_2000['Average grade']=df_2000['Grado promedio de escolaridad']
df_2005['Average grade']=df_2005['Grado promedio de escolaridad']
df_2010['Average grade']=df_2010['Grado promedio de escolaridad']
df_2020['Average grade']=df_2020['Grado promedio de escolaridad']

# Literacy 
df_2000['Illiteracy Population Age 15+']=df_2000['Población de 15 años y más analfabeta']/df_2000['Población de 15 años y más']
df_2005['Illiteracy Population Age 15+']=df_2005['Población de 15 años y más analfabeta']/df_2005['Población de 15 años y más']
df_2010['Illiteracy Population Age 15+']=df_2010['Población de 15 años y más analfabeta']/df_2010['Población de 15 años y más']
df_2020['Illiteracy Population Age 15+']=df_2020['Población de 15 años y más analfabeta']/df_2020['Población de 15 años y más']

# School assistance

df_2000['Population not attending school ( 15-24 )']=df_2000['Población de 15 a 24 años que no asiste a la escuela']/df_2000['Población de 15 a 24 años']
df_2005['Population not attending school ( 15-24 )']=(df_2005['Población de 15 a 24 años']-df_2005['Población de 15 a 24 años que asiste a la escuela'])/df_2005['Población de 15 a 24 años']
df_2010['Population not attending school ( 15-24 )']=(df_2010['Población de 15 a 17 años']-df_2010[ 'Población de 15 a 17 años que asiste a la escuela']
                                                    + df_2010['Población de 18 a 24 años']- df_2010['Población de 18 a 24 años que asiste a la escuela'])/( df_2010['Población de 15 a 17 años']+df_2010['Población de 18 a 24 años'])
df_2020['Population not attending school ( 15-24 )']=(df_2020['Población de 15 a 17 años']-df_2020[ 'Población de 15 a 17 años que asiste a la escuela']
                                                     +df_2020['Población de 18 a 24 años']- df_2020['Población de 18 a 24 años que asiste a la escuela'])/( df_2020['Población de 15 a 17 años']+df_2020['Población de 18 a 24 años'])

df_2000['Population not attending school ( 6-14 )']=df_2000['Población de 6 a 14 años que no asiste a la escuela']/df_2000['Población de 6 a 14 años']
df_2005['Population not attending school ( 6-14 )']=df_2005['Población de 6 a 14 años que no asiste a la escuela']/df_2005['Población de 6 a 14 años']
df_2010['Population not attending school ( 6-14 )']=(df_2010[ 'Población de 6 a 11 años que no asiste a la escuela']+ df_2010['Población de 12 a 14 años que no asiste a la escuela'])/( df_2010['Población de 6 a 11 años']+df_2010['Población de 8 a 14 años'])
df_2020['Population not attending school ( 6-14 )']=(df_2020[ 'Población de 6 a 11 años que no asiste a la escuela']+ df_2020['Población de 12 a 14 años que no asiste a la escuela'])/(df_2020['Población de 6 a 11 años']+df_2020['Población de 8 a 14 años'] )

# Year
df_2000['Year']=2000
df_2005['Year']=2005
df_2010['Year']=2010
df_2020['Year']=2020
```

```python
columns=['Entity','Average grade','Illiteracy Population Age 15+',
          'Population not attending school ( 6-14 )','Population not attending school ( 15-24 )','Year']
dataframes=[df_2000[columns],df_2005[columns],df_2010[columns],df_2020[columns]]
df=pd.concat(dataframes)
```

```python
df['Illiteracy Population Age 15+']=df['Illiteracy Population Age 15+'] * 100
df['Population not attending school ( 15-24 )']=df['Population not attending school ( 15-24 )']*100
df['Population not attending school ( 6-14 )']=df['Population not attending school ( 6-14 )']*100
```
